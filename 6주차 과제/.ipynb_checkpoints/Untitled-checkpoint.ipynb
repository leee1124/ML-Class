# 비지도 학습(비지도 변환, 군집)

### 비지도학습은 알고 있는 출력값이나 정보 없이 학습 알고리즘을 훈련하는 모든 종류의 머신러닝
### 입력 데이터만으로 데이터에서 지식을 추출
### 비지도 변환
- 데이터를 새롭게 표현하여 사람이나 머신러닝 알고리즘이 원래 데이터보다 쉽게 해석할 수 있도록 만드는 알고리즘
-  차원축소
  - 고차원 데이터의 특수성을 감소
  - 꼭 필요한 특성만 포함
  - 시각화를 위해 데이터셋을 2차원으로 변환
  - PCA, NMF, t-SNE, 오토인코더
- 데이터를 구성하는 단위나 성분 찾기
  - 많은 텍스트 문서에서 주제 찾기
  - 문서에서 이야기하는 주제는 무엇인지 찾고 학습
  - SNS에서 선거, 총기 규제, 팝스타 같은 주제의 토론을 추적
  - 토픽 모델링(LDA)

# 군집(clustering) 알고리즘
- 데이터를 비슷한 것끼리 묶음
  - 사진에 나타난 모든 얼굴을 추출하여 비슷한 얼굴로 그루핑
  - 객체 인식/객체수에는 관심없음
- K-평균, 병합, DBSCAN

# 비지도 학습의 어려움
- 알고리즘의 평가
  - 레이블이 없어서 유용성 평가가 어려움
  - 직접 확인
- 데이터를 잘 이해하기 위해 사용: 탐색적 분석 단계에서 사용
- 데이터의 전처리 단계
  - 알고리즘의 정확도 향상
  - 메모리와 시간을 절약하기 위해 전처리 단계 활용

# 데이터 변환 적용하기

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)
#유방암 데이터를 적재하고 훈련세트와 테스트 세트로 나눔.
#전처리 후에 만들어진 지도학습 모델을 평가하려면 훈련세트와 테스트세트로 나눠야함

print(X_train.shape)
print(X_test.shape)
#이 데이터셋에는 569개의 데이터 포인트가 있고 각 데이터 포인트는 30개의 측정 값으로 이루어짐
#샘플 426개를 훈련세트로, 143개를 테스트 세트로 나눔
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
#전처리가 구현된 파이썬 클래스를 임포트하고 객체 생성
```

```python
scaler.fit(X_train)
#fit 메서드에 훈련 데이터 적용
#MinMaxScaler의 fit 메서드는 훈련세트에 있는 특성마다 최솟값과 최댓값을 계산함
#스케일 객체는 fit메서드를 호출할 때 y_train은 사용하지 않음
```

```python
#데이터 변환
X_train_scaled = scaler.transform(X_train)
#스케일이 조정된 후 데이터셋의 속성을 출력
print("변환된 후 크기:", X_train_scaled.shape)
print("스케일 조정 전 특성별 최소값:\n", X_train.min(axis=0))
print("스케일 조정 전 특성별 최대값:\n", X_train.max(axis=0))
print("스케일 조정 후 특성별 최소값:\n", X_train_scaled.min(axis=0))
print("스케일 조정 후 특성별 최대값:\n", X_train_scaled.max(axis=0))
```

```python
#테스트 데이터 변환
X_test_scaled = scaler.transform(X_test)
#스케일이 조정된 후 테스트 데이터의 속성 출력
print("스케일 조정 후 특성별 최소값:\n", X_test_scaled.min(axis=0))
print("스케일 조정 후 특성별 최대값:\n", X_test_scaled.max(axis=0))
```

# 주성분 분석(PCA)
- 특성들이 통계적으로 상관관계가 없도록 데이터셋을 회전시키는 기술
- 데이터를 설명의 중요도에 따라 데이터 선택(종종 새로운 특성 중 일부만 선택됨)
- 고차원에서의 분산 유지하면서 차원을 줄임(주성분)
- y=Wx: 입력 x에 대한 W에 의한 선형변환

```python
mglearn.plots.plot_pca_illustration()
```

## PCA를 적용해 유방암 데이터셋 시각화
1. PCA가 가장 널리 사용되는 분야는 고차원 데이터셋의 시각화임
2. 유방암 데이터셋에는 산점도 행렬을 적용하기 어려움
3. 양성과 악성 두 클래스에 대해 각 특성의 히스토그램을 그리는 것이 쉬운 방법

```python
fig, axes = plt.subplots(15,2, figsize=(10,20))
malignant = cancer.data[cancer.target==0]
benign = cancer.data[cancer.target==1]

ax = axes.ravel()


for i in range(30):
    _, bins = np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:, i], bins=bins, color=mglearn.cm3(0), alpha=.5)
    ax[i].hist(benign[:,i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("특성 크기")
ax[0].set_ylabel("빈도")
ax[0].legend(["악성","양성"], loc="best")
fig.tight_layout
```

1. 이 그림은 유방암 데이터의 클래스별 특성 히스토그램으로 특정 간격(bin)에 얼마나 많은 데이터 포인트가 나타나는지 횟수를 센 것
2. 각 그래프는 히스토그램 두 개를 겹쳐놓은 것으로 초록색은 양성 클래스의 포인트를, 그리고 푸른색은 악성 클래스의 포인트를 나타냄
3. 이 그림은 특성들이 클래스별로 어떻게 분포되어 있는지를 알려주고, 이를 통해 어떤 특성이 양성과 악성 샘플을 구분하는 데 더 좋은지 가늠해볼 수 있음
4. 그러나 특성 간의 상호작용이나 이 상호작용이 클래스와 어떤 관련이 있는지는 전혀 알려주지 못함




```python
#PCA를 조정하기 전에 StandardScaler를 사용해 각 특성의 분산이 1이 되도록 데이터의 스케일 조정
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled=scaler.transform(cancer.data)
```

```python
from sklearn.decomposition import PCA
#데이터의 처음 두 갱 주성분만 유지
pca=PCA(n_components=2)
#유방암 데이터로 PCA 모델을 만듦
pca.fit(X_scaled)

#처음 두 개의 주성분을 사용해 데이터를 변환
X_pca = pca.transform(X_scaled)
print("원본 데이터 형태", str(X_scaled.shape))
print("축소된 데이터 형태:",str(X_pca.shape))
```

```python
#클래스를 색깔로 구분하여 처음 두 개의 주성분을 그래프로 나타냄
plt.figure(figsize=(8,8))
mglearn.discrete_scatter(X_pca[:,0],X_pca[:, 1], cancer.target)
plt.legend(["악성", "양성"], loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("첫 번째 주성분")
plt.ylabel("두 번째 주성분")
#처음 두 개의 주성분을 사용해 그린 유방암 데이터셋의 2차원 산점도
#악성의 분포가 넓고 주성분 2개로 잘 구분됨
```

```python
print("PCA 주성분 형태:", pca.components_.shape)
```

```python
print("PCA 주성분:", pca.components_)
```

```python
plt.matshow(pca.components_, cmap="viridis")
plt.yticks([0,1], ["첫 번쨰 주성분", "두 번째 주성분"])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)),
          cancer.feature_names, rotation=60, ha='left')
plt.xlabel("특성")
plt.ylabel("주성분")
```

유방암 데이터셋에서 찾은 처음 두 개의 주성분 히트맵

1. 데이터 사이의 상관관계로만 주성분을 찾음. 클래스 정보 고려하지 않음
2. 제 1성분(+값): 공통의 상호관계(한 특성이 커지면 다른 값들도 같이 높아짐)
3. 두번째 주성분: 부호가 섞여있어 축의 의미 설명이 어려움

## 고유얼굴(eigen face) 특성 추출
1. 특성추출: 원본 데이터보다 분석하기에 더 적합한 표현 추출
2. LFW(labeled faces in the wild)
3. 2000년대 초반 이후의 정치인, 가수, 배우, 운동선수들 얼굴
4. 총 45명(45개 클래스), 2537개 이미지, 87x65(5655)픽셀
5. 흑백(black/white), 스케일 조정
6. 얼굴 인식: 기존 DB클래스 중 어디에 속하는지 분류

```python
from sklearn.datasets import fetch_lfw_people
people = fetch_lfw_people(min_faces_per_person = 20, resize=0.7)
image_shape = people.images[0].shape

fig, axes = plt.subplots(2, 5, figsize=(15, 8),
                         subplot_kw={'xticks':(), 'yticks':()})
for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
print("people.images.shape:", people.images.shape)
print("클래스 개수:",len(people.target_names))
```

```python
#각 타깃이 나타난 횟수 계산
counts = np.bincount(people.target)
#타깃별 이름과 횟수 출력
for i, (count, name) in enumerate(zip(counts, people.target_names)):
    print("{0:25} {1:3}".format(name, count), end='    ')
    if(i+1)%3==0:
        print()
```
